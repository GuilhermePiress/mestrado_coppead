```python
# Coloca o símbolo de porcentagem nas colunas que estão em porcentagem
for col in df_acoes.columns:
    if any("%" in str(val) for val in df_acoes[col].values):
        df_acoes = df_acoes.rename(columns={col: col + "%"})
        
# Retira simbolo de '%' dos dados
# Retira simbolo de '-' isolado dos dados e preenche com nan
df_acoes_t1 = df_acoes.copy().replace(to_replace='%', value='', regex=True)\
                             .replace("-", np.nan)

# Fixa as casas decimais corretas para os dados númericos (coluna 2 em diante)
df_acoes_t1.iloc[:, 2:] = df_acoes_t1.iloc[:, 2:]\
                                     .apply(lambda x: pd.to_numeric(x.str.replace(".", "")
                                                                     .str.replace(",", ".")))
# Converte para os tipos corretos
df_acoes_t1[df_acoes_t1.columns[:2]] = df_acoes_t1[df_acoes_t1.columns[:2]].astype('object')
df_acoes_t1[df_acoes_t1.columns[2:]] = df_acoes_t1[df_acoes_t1.columns[2:]].astype('float')

# Fill nan values
df_acoes_t1.fillna(0, inplace=True)
#
df_acoes_t1.head()
```

# Clean bdr tickets csv

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df_bdrs = pd.read_csv("bdrs_stinvest.csv")
df_bdrs.head(10)
```

```python
df_bdrs.head()
```

```python
df_bdrs.info()
```

```python
# Coloca o símbolo de porcentagem nas colunas que estão em porcentagem
for col in df_bdrs.columns:
    if any("%" in str(val) for val in df_bdrs[col].values):
        df_bdrs = df_bdrs.rename(columns={col: col + "%"})
        
# Retira simbolo de '%' dos dados
# Retira simbolo de '-' isolado dos dados e preenche com nan
df_bdrs_t1 = df_bdrs.copy().replace(to_replace='%', value='', regex=True)\
                           .replace("-", np.nan)
# Fill nan values
df_bdrs_t1.fillna(0, inplace=True)

# Fixa as casas decimais corretas para os dados númericos (coluna 2 em diante)
df_bdrs_t1.iloc[:, 2:] = df_bdrs_t1.iloc[:, 2:]\
                                   .applymap(lambda x: pd.to_numeric(str(x).replace(".", "")
                                                                           .replace(",", ".")))
# Converte para os tipos corretos
df_bdrs_t1[df_bdrs_t1.columns[:2]] = df_bdrs_t1[df_bdrs_t1.columns[:2]].astype('object')
df_bdrs_t1[df_bdrs_t1.columns[2:]] = df_bdrs_t1[df_bdrs_t1.columns[2:]].astype('float')


df_bdrs_t1.head()
```

```python
df_bdrs_t1.info()
```

```python
df_bdrs_t1.to_csv('bdrs_b3_stinvest_tratado.csv', index=False)
```

# Clean FII tickets csv

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df_fii = pd.read_csv("FII_stinvest.csv", sep = ',')
# Retira dados com ano de 2023
df_fii = df_fii[df_fii['Anos'] != 2023].reset_index(drop = True)
df_fii.head(10)
```

```python
# Preenche dados com '-' por zero
df_fii_t1 = df_fii.copy().replace("-", np.nan)\
                         .fillna(0)

#Trata diferentes colunas
df_fii_t1 = df_fii_t1.applymap(lambda x: str(x).replace(".", "")
                                               .replace(",", ".")
                                               .replace(" ", ""))

def str_to_float(x):
    if x[-1] == 'K':
        return float(x[:-1]) * 1000
    elif x[-1] == 'M':
        return float(x[:-1]) * 1000000
    else:
        return float(x)

df_fii_t1.iloc[:, 2:] = df_fii_t1.iloc[:, 2:].applymap(str_to_float)    

# Converte para os tipos corretos
df_fii_t1[df_fii_t1.columns[:2]] = df_fii_t1[df_fii_t1.columns[:2]].astype('object')
df_fii_t1[df_fii_t1.columns[2:]] = df_fii_t1[df_fii_t1.columns[2:]].astype('float')
df_fii_t1['Valor patrimonial cota - (R$)'] = df_fii_t1['Valor patrimonial cota - (R$)']/100
df_fii_t1.head()
```

```python
df_fii_t1.to_csv('FII_b3_stinvest_tratado.csv', index=False)
```
